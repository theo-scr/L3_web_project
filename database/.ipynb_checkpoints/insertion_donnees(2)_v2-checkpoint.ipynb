# Insertion dans la table Represenation

```python
#Import

from sqlalchemy import create_engine
import cx_Oracle
import datetime
```

```python
# Connexion à la base de données

mydb = cx_Oracle.connect(
    'SCT2985A/29101999esg@//telline.univ-tlse3.fr:1521/etupre')
mycursor = mydb.cursor()
```

```python
# Requête permettant d'obtenir la liste des identifiants des figures qui sont en attente d'une représentation

sql_request = "SELECT idFigure FROM FIGURE MINUS (SELECT idFigure FROM REPRESENTATION)"
test = mycursor.execute(sql_request)
liste_id_figure = [row[0] for row in test]
```

```python
# Dans un premier temps, on renseigne l'identifiant de la figure dont nous souhaitons créer la représentation

print("Saisir l'identifiant de la figure dont vous souhaitez faire une représentation. \nAttention veuillez choisir un identifiant parmi la liste suivante : ",
      liste_id_figure, "\nTapez 0 pour annuler.")

id_fig = int(input("Identifiant figure: "))

# On vérifie que la saisie est valide
annulation = False

if (id_fig == 0):
    print("\nLa saisie de la reprensentation est ANNULEE\n")
    annulation = True

while(id_fig not in liste_id_figure and not annulation):
    print("\nERREUR - La saisie ne correspond pas à un identifiant valide. Veuillez recommencer. ")
    id_fig = int(input())
    if (id_fig == 0):
        print("\nLa saisie de la reprensentation est ANNULEE\n")
        annulation = True
```

```python
# Dictionnaire permettant de relier notre saisie avec ce que l'on souhaite insérer dans la base

dico = {"R": None, "Rg": "rouge", "Rs": "rose", "B": "bleu",
        "Ve": "vert", "Vi": "violet", "O": "orange", "A": "Annuler"}
```

```python
# Variables
annulation = False
j = 1

# Requête SQL
sql_request = "INSERT INTO Representation VALUES ( :1, :2, :3, :4)"


# On prévoit une valeur pour id_fig pour annuler
if (id_fig != 0):
    # Pour chaque case on demande quelle est la couleur à mettre
    while (j <= 10 and not annulation):
        i = 1
        while (i <= 10 and not annulation):
            couleur = input(
                "\nSaisir la couleur: R pour rien (case vide), Rg pour rouge, Rs pour rose, B pour bleu, Ve pour vert, Vi pour violet et O pour orange. Pour annuler tapez A. \nCase (%i,%i): " % (i, j))
            # On vérifie si la saisie est correct
            while (couleur not in dico):
                print(
                    "\nERREUR - La saisie ne correspond pas à une couleure valide. Veuillez recommencer. ")
                couleur = input()

            # On vérifie si on est dans le cas d'annulation
            if (couleur == "A"):
                print("\nLa saisie de la reprensentation est ANNULEE\n")
                annulation = True
            # Si ce n'est pas le cas on insère
            else:
                values = list([id_fig, i, j, dico[couleur]])
                mycursor.execute(sql_request, values)
            i += 1
        j += 1
    # Pour finir on vérifie si on est sorti des boucles parce qu'on a demandé d'annuler ou si parce qu'on a rempli toutes les cases.
    # Si on a annulé, on rollback
    if (annulation):
        mydb.rollback()
    # Si on a rempli toutes les cases, on commit
    else:
        print("\nFigure terminée. Tout s'est bien passé.")
        mydb.commit()
```
